```python
import xarray as xr
import numpy as np
import os
import arrow
import pandas as pd
import metpy.calc as mpcalc
from metpy.units import units
```

```python
fcHour_list = np.array(list(range(0, 72+1, 3)) + list(range(78, 240+1, 6)))
fcHour_list_target = np.array(range(0, 240+1, 1))
```

```python
dir_path = 'H:/github/python/seafog/data/CFdata/concat/'
file_name = '201810-202205_lon110.25_lat20.125.sstk.nc'
ds = xr.open_dataset(f'{dir_path}{file_name}')
```

```python
iHour_target = 5
if(iHour_target in fcHour_list): 
    print('已存在时间')
else:
    # 寻找最近的两个值
    hour_distance = fcHour_list - iHour_target
    left_shift  = np.max(hour_distance[hour_distance<0])
    right_shift = np.min(hour_distance[hour_distance>0])
    varname = f'{"sstk"}{iHour_target:0>3d}'
    left_hour = iHour_target + left_shift
    right_hour = iHour_target + right_shift 
    varname_left = f'{"sstk"}{left_hour:0>3d}'
    varname_right = f'{"sstk"}{right_hour:0>3d}'
    with xr.set_options(keep_attrs=True):
        ds[varname] = (ds[varname_left]*(right_hour-iHour_target) + ds[varname_right]*(iHour_target-left_hour))/(right_hour-left_hour)
```

```python
with xr.set_options(keep_attrs=True):
    ds['sstk005'] = (ds['sstk003']*(right_hour-iHour_target) + ds['sstk006']*(iHour_target-left_hour))/(right_hour-left_hour)
print(ds['sstk005'])
```
